```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'],categories = pd.factorize(df['species'])
df.head()
```

Applying elbow Method

```python
k_rng = range(1, 10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
import matplotlib.pyplot as plt

# Assume k_rng and sse are provided as lists
# Here we'll create sample data for demonstration purposes
k_rng = [1, 2, 3, 4, 5]
sse = [100, 80, 60, 40, 20]

# Plotting
plt.plot(k_rng, sse, marker='o', linestyle='-')

# Labeling the axes
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors')

# Title
plt.title('Elbow Method: Sum of Squared Errors vs. Number of Clusters')

# Display the plot
plt.grid(True)
plt.show()
```

Applying KMeans Algorithm

```python
km = KMeans(n_clusters=3, random_state=0)
y_predicted= km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']= y_predicted
df.head(150)
```

MEASUREMENT OF ACCURACY

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assume true_labels and predicted_labels are the true and predicted labels, respectively
# Here we'll create sample labels for demonstration purposes
true_labels = ['Setosa', 'Versicolor', 'Virginica', 'Setosa', 'Virginica']
predicted_labels = ['Setosa', 'Versicolor', 'Versicolor', 'Setosa', 'Virginica']

# Compute confusion matrix
unique_labels = np.unique(true_labels)
num_classes = len(unique_labels)
confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

for true_label, predicted_label in zip(true_labels, predicted_labels):
    true_index = np.where(unique_labels == true_label)[0][0]
    predicted_index = np.where(unique_labels == predicted_label)[0][0]
    confusion_matrix[true_index, predicted_index] += 1

# Plot confusion matrix
plt.imshow(confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, unique_labels, rotation=45)
plt.yticks(tick_marks, unique_labels)

# Fill matrix with values
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, str(confusion_matrix[i, j]), ha='center', va='center', color='white')

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()
```
